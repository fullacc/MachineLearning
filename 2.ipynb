```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def func(x):
    return (np.sin(x / 5)) *( np.exp(x / 10)) + 5 * (np.exp(-x / 2))
def y(w):
    y = np.zeros(len(x))
    for i in range(len(x)):
        for j in range(len(w)):
            y[i] += w[j] * (x[i] ** j)
    return y

def coef(arr):
    v = []
    c = [[0 for j in range(len(arr))] for i in range(len(arr))]
    for i in range(len(arr)):
        v.append(func(arr[i]))
        for j in range(len(arr)):
            c[i][j] = arr[i]**j
    return np.linalg.solve(c, v)
x = np.arange(1, 15, 0.1)
plt.plot(x, func(x))
```

```python
A = np.matrix('1 1; 1 15')
B = np.array([func(1), func(15)])
w_1 = np.polynomial.polynomial.Polynomial(np.linalg.solve(A, B))

y_1 = w_1(x)
plt.plot(x, func(x), '-', x, y_1, '-')
```

```python
x_2 = np.array([1, 8, 15])
line_2 = np.polyfit(x_2, func(x_2), 2)
w_2 = np.poly1d(line_2)
y_2 = w_2(x)
plt.plot(x, func(x), '-', x, y_2, '-')
```

```python
w3 = coef([1, 4, 10, 15])
y3 = y(w3) 
plt.plot(x, y3)
plt.plot(x, list(map(func, x)))
```
